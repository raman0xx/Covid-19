# 14. Plotting bar chart : How many orders (per month)?


```python
plt.xlabel('Month')
plt.ylabel('Numbers of orders')
plt.title('Number of orders for different months')
df.Month.value_counts().plot(kind='bar',cmap='hot')
plt.grid
```

# 15. Plotting bar chart : How many orders (per year)?

```python
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.title('Number of orders for different Year')
df.Year.value_counts().plot(kind='bar',cmap='cool')
plt.grid
```

# 16. Plotting bar chart : How many orders (per hour)?
